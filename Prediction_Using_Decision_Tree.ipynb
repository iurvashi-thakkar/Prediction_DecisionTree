```python
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y=df['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=1)
```

Generation of Decision Tree Model

```python
df_first = DecisionTreeClassifier()
```

```python
df_first.fit(X_train, y_train)
```

```python
Y_prediction=df_first.predict(X_test)
```

confusion matrix is used to evaluate quality of output classifier.The diagonal elements show what howmany prediction values actually match with actual values.Higher the number of diagonal the indicate higher number of correct prediction which says classifier's good quality 

```python
confusion_matrix(y_test,Y_prediction)
```

Classification report

```python
print(classification_report(y_test, Y_prediction))
```

```python
# List of values to try for max_depth:
max_depth_range = list(range(1, 6))
# List to store the accuracy for each value of max_depth:
accuracy = []
for depth in max_depth_range:
    
    df = DecisionTreeClassifier(max_depth = depth, 
                             random_state = 0)
    df.fit(X_train, y_train)
    y_pred = df.predict(X_test)
    score = df.score(X_test, y_test)
    accuracy.append(score)
```
